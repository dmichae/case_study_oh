```python
facilities_df = pd.read_csv('Facilities.csv')
provider_df = pd.read_csv('Provider Demographics.csv')
dem_df = pd.read_csv('Provider-Facility Affilitiated Cl.csv')

print(facilities_df.shape)
print(provider_df.shape)
print(dem_df.shape)
```

```python
# Filtering Specialty
filtered_provider = provider_df[(provider_df['Specialty'].isin(['INTERNAL MEDICINE', 'UROLOGY', 'ANESTHESIOLOGY']))]
print(filtered_provider.shape)
```

```python
# Provider Demographics
print("Basic Data Structure:")
print(dem_df.info())

# Summary Statistics
print("\nSummary Statistics:")
print(dem_df.describe())

# Check for missing values
print("\nMissing Values:")
print(dem_df.isnull().sum())

dem_df.head(5)
```

```python
# Merging facilities, provider, and provider demographics
provider_dem = pd.merge(filtered_provider, dem_df, left_on='NPI', right_on='Clinician NPI')
merged_data = pd.merge(provider_dem, facilities_df, left_on='Facility NPI', right_on='npi')
print(provider_dem.shape)
print(merged_data.shape)
merged_data.head(5)
```

```python
# Cleaning
columns_to_drop = [
    'incorporate_date', 'incorporate_state', 'organization_type_structure',
    'organization_other_type_text', 'proprietarynonprofit_flag',
    'addressline1', 'addressline2', 'city', 'state', 'zip_code',
    'practice_location_type', 'location_other_type_text',
    'subgroup_general_flag', 'subgroup_acute_care_flag',
    'subgroup_alcohol_drug_flag', 'subgroup_children_hospital_flag',
    'subgroup_longterm_flag', 'subgroup_psychiatric_flag',
    'subgroup_rehabilitation_flag', 'subgroup_shortterm_flag',
    'subgroup_swingbed_approved_flag', 'subgroup_psychiatric_unit_flag',
    'subgroup_rehabilitation_unit_flag', 'subgroup_speciality_hospital_flag',
    'subgroup_other_flag', 'subgroup_other_text', 'subgroup_reh_conversion_flag',
    'subgroup_reh_conversion_date', 'subgroup_cah_ccn', 'Address Line 2', 'provider_type_text',
    'ccn', 'associate_id', 'enrollment_id', 'provider_type_code', 'Zip', 'Phone', 'Address', 
    'Provider Credentials'
]

merged_data = merged_data.drop(columns=columns_to_drop, axis=1)
print(merged_data.shape)
merged_data.head(5)
```

```python
# Group by Healthcare Organization and Specialty for total claims, total patients, and unique provider count
detailed_summary = merged_data.groupby(['Healthcare Organization Name', 'Specialty']).agg(
    total_claims=('total_claims', 'sum'),
    total_patients=('total_patients', 'sum'),
    unique_providers_count=('Provider Name', 'nunique')
).reset_index()

# Printing out the DataFrame
print(detailed_summary)
```

```python
clinician_summary = merged_data.groupby(['Clinician NPI', 'Specialty'])['total_claims', 'total_patients'].sum()#.sort_values(ascending=False)
clinician_summary['zero_claims_flag'] = clinician_summary['total_claims'] == 0
clinician_summary['zero_patients_flag'] = clinician_summary['total_patients'] == 0
clinician_summary
```

```python
zero_claims_count = (clinician_summary['total_claims'] == 0).sum()
print(f"Number of clinicians with zero claims: {zero_claims_count}")
zero_claims_count = (clinician_summary['total_patients'] == 0).sum()
print(f"Number of clinicians with zero patients: {zero_claims_count}")
```

```python
# Zero Claims Filter
zero_claims_by_specialty = clinician_summary[clinician_summary['total_claims'] == 0].groupby('Specialty').size()
print(zero_claims_by_specialty)
```

```python
clinician_summary.head(5)
```

```python
# Reset the index so 'Clinician NPI' and 'Specialty' become columns
clinician_summary_reset = clinician_summary.reset_index()

# Create the flag DataFrame
flag_df = clinician_summary_reset[['Clinician NPI', 'total_claims']].copy()
flag_df['zero_claims_flag'] = flag_df['total_claims'] == 0

# Reduce to just the NPI and the flag, and remove duplicates
flag_df = flag_df[['Clinician NPI', 'zero_claims_flag']].drop_duplicates()
flag_df.shape
```

```python
final_data = merged_data.merge(flag_df, on='Clinician NPI', how='left')
print(final_data.shape)
# Drop clinicians with zero claims
final_data = final_data[final_data['zero_claims_flag'] == False]

# Drop the 'zero_claims_flag' column if no longer needed
final_data.drop('zero_claims_flag', axis=1, inplace=True)

print(final_data.shape)
final_data.head(5)
```

```python
# Group by Healthcare Organization and Specialty for total claims, total patients, and unique provider count
detailed_summary = final_data.groupby(['Healthcare Organization Name', 'Specialty']).agg(
    total_claims=('total_claims', 'sum'),
    total_patients=('total_patients', 'sum'),
    unique_providers_count=('Provider Name', 'nunique')
).reset_index()

# Printing out the DataFrame
print(detailed_summary)
```

```python
# Convert 'quarter' and 'year' to numerical values
final_data['quarter'] = final_data['quarter'].astype(int)
final_data['year'] = final_data['year'].astype(int)

# Create a new column for numerical representation of 'quarter_year'
final_data['time_period'] = final_data['year'] * 100 + final_data['quarter']

# Convert 'time_period' to string if it's not already
final_data['time_period'] = final_data['time_period'].astype(str)
```

```python
# Claims trends by facility by specialty
sns.set_style("white")
plt.rcParams['lines.linewidth'] = 3
# Define a color palette for consistency
palette = sns.color_palette("pastel")
agg_data = final_data.groupby(['Healthcare Organization Name', 'Specialty', 'time_period'])[['total_claims', 'total_patients']].sum().reset_index()
specialties = agg_data['Specialty'].unique()

for specialty in specialties:
    specialty_data = agg_data[agg_data['Specialty'] == specialty]

    # Pivot the data for plotting
    agg_claims = specialty_data.pivot_table(index='time_period', columns='Healthcare Organization Name', values='total_claims')
    agg_patients = specialty_data.pivot_table(index='time_period', columns='Healthcare Organization Name', values='total_patients')

    # Define the figure and axes
    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(20, 8))
    
    # Plot for Total Claims
    agg_claims.plot(ax=ax[0], color=palette)
    ax[0].set_title(f'Total Claims by Specialty - {specialty}', fontsize=16)
    ax[0].set_ylabel('Total Claims', fontsize=14)
    ax[0].set_xlabel('Year - Quarter', fontsize=14)
    ax[0].tick_params(axis='x', rotation=45)
    ax[0].tick_params(axis='both', labelsize=12)

    # Plot for Total Patients
    agg_patients.plot(ax=ax[1], color=palette)
    ax[1].set_title(f'Total Patients by Specialty - {specialty}', fontsize=16)
    ax[1].set_ylabel('Total Patients', fontsize=14)
    ax[1].set_xlabel('Year - Quarter', fontsize=14)
    ax[1].tick_params(axis='x', rotation=45)
    ax[1].tick_params(axis='both', labelsize=12)

    plt.tight_layout()
    plt.savefig(f"count_of_{specialty.replace(' ', '_')}_by_claims and patients with cleveland.png", dpi=300)
    plt.show()

    # Now, for the second set of plots, exclude the Cleveland Clinic
    specialty_data_excl_cc = specialty_data[~specialty_data['Healthcare Organization Name'].str.contains("CLEVELAND CLINIC")]

    # Pivot the data for plotting without Cleveland Clinic
    agg_claims_excl_cc = specialty_data_excl_cc.pivot_table(index='time_period', columns='Healthcare Organization Name', values='total_claims')
    agg_patients_excl_cc = specialty_data_excl_cc.pivot_table(index='time_period', columns='Healthcare Organization Name', values='total_patients')

    # Define the figure and axes
    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(20, 8))
    
    # Plot for Total Claims excluding Cleveland Clinic
    agg_claims_excl_cc.plot(ax=ax[0], color=palette)
    ax[0].set_title(f'Total Claims by Specialty Excluding Cleveland Clinic - {specialty}', fontsize=16)
    ax[0].set_ylabel('Total Claims', fontsize=14)
    ax[0].set_xlabel('Year - Quarter', fontsize=14)
    ax[0].tick_params(axis='x', rotation=45)
    ax[0].tick_params(axis='both', labelsize=12)

    # Plot for Total Patients excluding Cleveland Clinic
    agg_patients_excl_cc.plot(ax=ax[1], color=palette)
    ax[1].set_title(f'Total Patients by Specialty Excluding Cleveland Clinic - {specialty}', fontsize=16)
    ax[1].set_ylabel('Total Patients', fontsize=14)
    ax[1].set_xlabel('Year - Quarter', fontsize=14)
    ax[1].tick_params(axis='x', rotation=45)
    ax[1].tick_params(axis='both', labelsize=12)

    plt.tight_layout()
    plt.savefig(f"count_of_{specialty.replace(' ', '_')}_by_claims and patients.png", dpi=300)
    plt.show()
    
```

```python
# Total Claims by Total Patients by Specialty
unique_specialties = final_data['Specialty'].unique()

# Set the style for the plots
sns.set_style("white")
sns.set_context("talk")  # 'talk' context is suitable for presentations

# Creating separate scatter plots for each specialty
for specialty in unique_specialties:
    # Filter data for the current specialty
    specialty_data = merged_data[merged_data['Specialty'] == specialty]

    # Calculate the limits for the current specialty
    max_claims_limit = specialty_data['total_claims'].max()
    max_patients_limit = specialty_data['total_claims'].max()
    min_claims_limit = specialty_data['total_claims'].min()
    min_patients_limit = specialty_data['total_claims'].min()

    # Scatter plot for the current specialty
    plt.figure(figsize=(12, 8))
    sns.scatterplot(data=specialty_data, x='total_patients', y='total_claims', hue='total_claims', s=100)

    plt.title(f'Total Patients vs Total Claims in {specialty}', fontsize=16)
    plt.xlabel('Total Patients', fontsize=14)
    plt.ylabel('Total Claims', fontsize=14)
    plt.legend(title='Provider Name', bbox_to_anchor=(1.05, 1), loc=2)

    # Set axis scales for the current plot
    plt.xlim(min_claims_limit, max_claims_limit)
    plt.ylim(min_patients_limit, max_patients_limit)

    plt.tight_layout()
    plt.savefig(f"claims_vs_patients_{specialty.replace(' ', '_')}.png", dpi=300)
    plt.show()
```

```python
# Overall Trends by specialty
print("\nTrends Over Time:")
for specialty in ['INTERNAL MEDICINE', 'UROLOGY', 'ANESTHESIOLOGY']:
    # Filter data for the current specialty
    specialty_data = final_data[final_data['Specialty'] == specialty]

    # Group by 'time_period' and sum only 'total_claims' and 'total_patients'
    grouped_data = specialty_data.groupby('time_period')[['total_claims', 'total_patients']].sum()

    plt.figure(figsize=(10, 4))

    # Plotting the summed data
    sns.lineplot(data=grouped_data, x=grouped_data.index, y='total_claims', label='Total Claims')
    sns.lineplot(data=grouped_data, x=grouped_data.index, y='total_patients', label='Total Patients')

    plt.title(f'Trends for {specialty}')
       # Renaming the y-axis
    plt.ylabel('Total Count')
    plt.xlabel('Year - Quarter')
    plt.legend()
    plt.xticks(rotation=45)
    plt.savefig(f"overview_{specialty.replace(' ', '_')}.png", dpi=300)
    plt.show()
```

```python
# Top 15 providers by specialty
for specialty in unique_specialties:
    # Filter data for the current specialty
    specialty_data = final_data[final_data['Specialty'] == specialty]

    # Get the top 15 providers for the current specialty
    top_providers = specialty_data.groupby('Provider Name').sum().nlargest(15, 'total_claims').reset_index()

    # Plotting directly with improved layout
    plt.figure(figsize=(10, 8))  # Adjust the size as needed for clarity
    barplot = sns.barplot(data=top_providers, x='Provider Name', y='total_claims', ci=None)
    plt.title(f'Top 15 Providers in {specialty} by Claims', fontsize=16)
    plt.xticks(rotation=45, fontsize=12)
    plt.yticks(fontsize=12)
    plt.ylabel('Total Number of Claims', fontsize=14)
    plt.xlabel('Provider Name', fontsize=14)

    # Highlight the highest bar to draw attention
    highest_claim = top_providers['total_claims'].max()
    for index, row in top_providers.iterrows():
        if row['total_claims'] == highest_claim:
            barplot.patches[index].set_facecolor('#ff9999')

    # Remove top and right borders for a cleaner look
    sns.despine()

    # Export the plot with high DPI for presentation clarity
    plt.tight_layout()
    plt.savefig(f"top_15_{specialty.replace(' ', '_')}_by_claims.png", dpi=300)
    plt.show()
```

```python
# Box plot per specialty
unique_specialties = final_data['Specialty'].unique()
grouped_data = final_data.groupby(['Specialty', 'Provider Name'])[['total_claims', 'total_patients']].sum().reset_index()
for specialty in unique_specialties:
    # Filter data for the current specialty
    specialty_data = grouped_data[grouped_data['Specialty'] == specialty]

    # Calculate average utilization for the current specialty
    avg_claims = specialty_data['total_claims'].mean()
    avg_patients = specialty_data['total_patients'].mean()

    # Add columns for deviation from average in the filtered data
    specialty_data['claims_deviation'] = specialty_data['total_claims'] - avg_claims
    specialty_data['patients_deviation'] = specialty_data['total_patients'] - avg_patients

    # Plotting for the current specialty
    plt.figure(figsize=(9, 7))
    sns.boxplot(x='Specialty', y='claims_deviation', data=specialty_data)
    plt.title(f'Claims Deviation from Average - {specialty}')
    plt.xlabel('Specialty')
    plt.ylabel('Claims Deviation')
    plt.xticks(rotation=45)
     # Annotate with the mean value
    plt.text(0.4, avg_claims, f'Mean: {avg_claims:.2f}', verticalalignment='center', horizontalalignment='right', fontsize=12)
    plt.savefig(f"boxplots_claims{specialty.replace(' ', '_')}.png", dpi=300)
    plt.show()
```

```python
# Calculate utilization ratio (total_claims / total_patients)
final_data['utilization_ratio'] = final_data['total_claims'] / final_data['total_patients']

# Display the DataFrame
print(final_data[['NPI', 'Provider Name', 'Specialty', 'Clinician NPI', 'time_period', 'utilization_ratio']])
```
